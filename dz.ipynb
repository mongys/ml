```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
```

```python
# Загрузка и предобработка данных
(x_train, y_train), (x_val, y_val) = mnist.load_data()
x_train = x_train / 255.0
x_val = x_val / 255.0

# Преобразование меток в one-hot encoding
y_train = to_categorical(y_train, 10)
y_val = to_categorical(y_val, 10)
```

*Модель без скрытых слоев (линейная модель)*

Accuracy без скрытых слоев: 0.9266999959945679

```python
# Модель без скрытых слоев (линейная модель)
model = Sequential([
    Flatten(input_shape=(28, 28)),  # Преобразование 28x28 изображений в плоский вектор
    Dense(10, activation='softmax') # Прямое преобразование входа в 10 классов
])

# Компиляция модели
model.compile(optimizer=Adam(),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Обучение модели
model.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=10, batch_size=32)

# Оценка модели на валидационном наборе данных
val_loss, val_accuracy = model.evaluate(x_val, y_val)
print(f"Accuracy без скрытых слоев: {val_accuracy}")
```

*Модель с двумя скрытыми слоями на 128 нейронов*    

Accuracy с двумя скрытыми слоями: 0.9785000085830688


```python
# Модель с двумя скрытыми слоями по 128 нейронов
model_with_hidden = Sequential([
    Flatten(input_shape=(28, 28)),       # Преобразование 28x28 изображений в плоский вектор
    Dense(128, activation='relu'),       # Первый скрытый слой
    Dense(128, activation='relu'),       # Второй скрытый слой
    Dense(10, activation='softmax')      # Выходной слой
])

# Компиляция модели
model_with_hidden.compile(optimizer=Adam(),
                          loss='categorical_crossentropy',
                          metrics=['accuracy'])

# Обучение модели
model_with_hidden.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=10, batch_size=32)

# Оценка модели на валидационном наборе данных
val_loss, val_accuracy_hidden = model_with_hidden.evaluate(x_val, y_val)
print(f"Accuracy с двумя скрытыми слоями: {val_accuracy_hidden}")
```
